```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
read_from_path = 'depth_ex.png'
img = cv2.imread(read_from_path)[:,:,::-1]
plt.title('depth (color)')
plt.imshow(img, cmap='gray')
plt.show()
img = cv2.imread(read_from_path, cv2.IMREAD_GRAYSCALE)/255
cv2.circle(img, (int(0), int(0)), int(100), (1, 0, 0), -1)
plt.title('depth (gray)')
plt.imshow(img, cmap='gray')
plt.show()
height = img.shape[0]
width = img.shape[1]
height_cut = int(height/3)
width_cut = int(width/3)
directions = {
    'left':(0, 1)
    ,'right':(2, 1)
    ,'up':(1, 0)
    ,'down':(1, 2)
    ,'forward':(1, 1)
       }
for direction in directions:
    x = directions[direction][0]
    y = directions[direction][1]
    img_cut = img[y*height_cut:(y+1)*height_cut, x*width_cut:(x+1)*width_cut]
    plt.title(direction)
    plt.imshow(img_cut, cmap='gray')
    plt.show()
    reward = 1 - np.mean(img_cut)
    print('reward', reward)
    directions[direction] = reward
```
